## Seaborn

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

### Correlation with Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

   ### Jointplot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg' )
```

### Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

### Distplot

```python
sns.distplot(df['tip'])
```

## Categorical Plots
### Countplot

```python

sns.countplot('sex', data=df)
```

```python
sns.countplot(y='day', data=df)
```

```python
sns.barplot(x='sex', y='tip', data=df)
```

### BoxPlot

```python
sns.boxplot('sex', 'total_bill', data=df, palette='rainbow')
```

```python
sns.boxplot(data=df, orient='h')
```

```python
sns.boxplot(x='total_bill', y='day', hue='smoker', data=df)
```

### Violin Plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
