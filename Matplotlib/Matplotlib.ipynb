```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

### Line Plot

```python
# Plotting the curve
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine wave')
plt.show()
```

```python
# Parabola
x = np.linspace(-10,10,20)
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'go')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'b.')
plt.plot(x, np.cos(x), 'r+')
```

### BarPlot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Spanish', 'Latin']
people = [100,50,150,40,70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel("NO. OF PEOPLE")
plt.show()
```

### Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'German', 'Spanish', 'Latin']
people = [100,50,150,40,70]
ax.pie(people, labels=languages, autopct='%1.1f%%' )
plt.show()
```

### Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
# useful in clusterring applications
```

### 3-D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
